### วิธีการคำนวณภาษี
1. ตรวจสอบรายได้ต่อปีของตัวเอง
2. คำนวณหารายได้ต่อปีสุทธิ โดยนำรายได้ต่อปีไปลดหย่อยภาษี (หักออก 60,000 บาท)
3. ตรวจสอบว่ารายได้ต่อปีสุทธิอยู่ที่ขั้นไหนของตาราง
4. ดูค่า จุดเริ่มต้นของขั้นนั้น, อัตราภาษีของขั้นนั้น และ ภาษีสูงสุดในแต่ละขั้นรายได้ของขั้นนั้น
5. ภาษี = (รายได้ต่อปีสุทธิ - จุดเริ่มต้นของขั้นนั้น) x อัตราภาษีของขั้นนั้น + ภาษีสูงสุดในแต่ละขั้นรายได้

**ตารางอัตราภาษีเงินได้บุคคลธรรมดาสำหรับปีภาษี 2560 ที่จะต้องยื่นรายการในปี พ.ศ.2561 เป็นต้นไป**



```python
salary = int(input("ใส่รายได้ต่อเดือน:"))                    # รายได้ต่อเดือน
income_per_year = salary * 12                                # รายได้ต่อปี   

print(f"รายได้ต่อเดือน = {salary} ,รายได้ต่อปี  = {income_per_year}")    # print ข้อมูลรายได้ต่อเดือนและรายได้ต่อปี
income_per_year -= 60_000                                           # รายได้ต่อปีสุทธิ เมื่อลดหย่อนภาษีแล้ว (รายได้ต่อปี - 60,000)

tax_rate = 0                            # ประกาศตัวแปรเก็บค่า อัตราภาษี
tax = 0                                 # ประกาศตัวแปรเก็บค่า ภาษีที่ต้องเสีย

# สูตรการคำนวณภาษี
# ภาษี = (รายได้ต่อปีสุทธิ - จุดเริ่มต้นของขั้นนั้น) x อัตราภาษีของขั้นนั้น + ภาษีสูงสุดในแต่ละขั้นรายได้

# คำถาม: ทำไมจากเงื่อนไขต้องเริ่มจากขั้นที่สูงที่สุด ?
# ตอบ : เพื่อป้องกัน input ที่ติดลบ
#       ในความเป็นจริงแล้ว input ที่ user ใส่ได้คือจำนวนเต็ม หมายความว่า user สามารถใส่ค่าติดลบได้
#       อาจจะแปลว่า user ติดหนี้! แน่นอนว่าเขาไม่จำเป็นต้องเสียภาษีด้วยซ้ำ
#       จากโค้ด เริ่มเงื่อนไขที่ขั้นที่สูงที่สุด (5,000,000) และให้เงื่อนไขสุดท้ายหรือ else ทำงานเมื่อ
#       user ใส่รายได้ที่อยู่ในช่วง 0 - 150,000 และรายได้ติดลบ

if income_per_year > 5_000_000:         # รายได้ต่อปีสุทธิมากกว่า 5,000,000 บาท
    # อัตราภาษี 35%
    tax_rate = 35                       
    tax = (income_per_year - 5_000_000) * (tax_rate / 100) + 1_265_000
    
elif income_per_year > 2_000_000:       # รายได้ต่อปีสุทธิมากกว่า 2,000,000 บาท
    # อัตราภาษี 30%
    tax_rate = 30
    tax = (income_per_year - 2_000_000) * (tax_rate / 100) + 365_000
    
elif income_per_year > 1_000_000:       # รายได้ต่อปีสุทธิมากกว่า 1,000,000 บาท
    # อัตราภาษี 25%
    tax_rate = 25
    tax = (income_per_year - 1_000_000) * (tax_rate / 100) + 11500
    
elif income_per_year > 750_000:         # รายได้ต่อปีสุทธิมากกว่า 750,000 บาท
    # อัตราภาษี 20%
    tax_rate = 20
    tax = (income_per_year - 750_000) * (tax_rate / 100) + 65000
    
elif income_per_year > 500_000:         # รายได้ต่อปีสุทธิมากกว่า 500,000 บาท
    # อัตราภาษี 15%
    tax_rate = 15
    tax = (income_per_year - 500_000) * (tax_rate / 100) + 27500
    
elif income_per_year > 300_000:         # รายได้ต่อปีสุทธิมากกว่า 300,000 บาท
    # อัตราภาษี 10%
    tax_rate = 10
    tax = (income_per_year - 300_000) * (tax_rate / 100) + 7500
    
elif income_per_year > 150_000:         # รายได้ต่อปีสุทธิมากกว่า 150,000 บาท
    # อัตราภาษี 5%
    tax_rate = 5
    tax = (income_per_year - 150_000) * (tax_rate / 100)
    
else:                                   # นอกเหนือจากนี้
    # อัตราภาษี 0 ไม่ต้องเสียภาษี
    tax_rate = 0
    tax = 0

print("อัตราภาษี", tax_rate)             # print อัตราภาษี
print("เสียภาษี =",tax)                  # print ภาษีที่ต้องจ่าย
```

# WorkShop Day 2

**ตารางราคา**



```python
def noodle_price(sen, nam, nuea, size):         # function คำนวณราคารายการก๋วยเตี๋ยว
    price = 0                                   # ประกาศตัวแปรเก็บ ราคาทั้งหมด ของรายการ

    # ประเภทเส้นก๋วยเตี๋ยว
    if sen == "เส้นเล็ก" or sen == "วุ้นเส้น":
        price += 0
    elif sen == "เส้นหมี่ขาว":
        price += 10
    elif sen == "เส้นใหญ่":
        price += 15
    elif sen == "เส้นหมี่เหลือง":
        price += 20

    # หมวดน้ำซุปก๋วยเตี๋ยว
    if nam == "แห้ง" or nam == "น้ำตก":
        price += 0
    elif nam == "น้ำใส":
        price += 10
    elif nam == "ต้มยำ" or nam == "เย็นตาโฟ":
        price += 20

    # หมวดหมูตุ๋น / เนื้อตุ๋น
    if nuea == "หมูตุ๋น":
        price += 10
    elif nuea == "เนื้อตุ๋น":
        price += 20

    # ธรรมดา / พิเศษ
    if size == "ธรรมดา":
        price += 0
    elif size == "พิเศษ":
        price += 10

    return price                                    # return ราคาทั้งหมดของรายการ
# ----------------------------------------------------------------------------------------------------

def drink_price(drink, sweet, cup):                 # function คำนวณราคารายการเครื่องดืม
    price = 0                                       # ประกาศตัวแปรเก็บ ราคาทั้งหมด ของรายการ

    # ประเภทเคครื่องดื่ม
    if drink == "ชาไทย" or drink == "โอเลี้ยง":   # ชาไทย หรือ โอเลี้ยง
        price += 0
    elif drink == "นมชมพู":                    # นมชมพู
        price += 20
    elif drink == "น้ำเปล่า":                       # น้ำเปล่า
        price += 10

    # ขนาดแก้ว
    if cup == "เล็ก":                           # แก้วเล็ก
        price += 0
    elif cup == "ใหญ่":                         # แก้วใหญ่
        price += 10

    return price, sweet                             # return ราคาทั้งหมดของรายการ และ ระดับบความหวาน
# ----------------------------------------------------------------------------------------------------

def main():                                         # function หลักที่จะรับออเดอร์
    orders = []                                     # ออเดอร์เก็บรายการอาหารที่สั่ง

    while True:                                     # วน loop infinity
        print("\nเลือกประเภทรายการ: (1) ก๋วยเตี๋ยว / (2) เครื่องดื่ม / (0) เพื่อปิดออเดอร์")
        choice = input("โปรดพิมพ์ตัวเลข: ")      # เลือกประเภทรายการด้วยตัวเลข
        
        if choice == "0":                     # ถ้าเลือก 0
            break                             # หยุด loop ปิดออเดอร์

        elif choice == "1":                   # ถ้าเลือก 1
            print("🍜 เมนูก๋วยเตี๋ยว")
            
            print("เส้น: เส้นเล็ก/วุ้นเส้น/เส้นหมี่ขาว/เส้นใหญ่/เส้นหมี่เหลือง")
            sen = input("เลือกเส้นก๋วยเตี๋ยว: ")
            
            print("น้ำซุป: แห้ง/น้ำใส/น้ำตก/ต้มยำ/เย็นตาโฟ")
            nam = input("เลือกน้ำซุป: ")
            
            print("เนื้อ: หมูตุ๋น/เนื้อตุ๋น")
            nuea = input("เลือกเนื้อ: ")
            
            size = input("เลือกขนาด (พิเศษ/ธรรมดา): ")
            
            price = noodle_price(sen, nam, nuea, size)                      # เรียก function คำนวณราคาก๋วยเตี๋ยว
            
            orders.append(f"ก๋วยเตี๋ยว {sen} {nam} {nuea} {size} = {price}฿")  # เพิ่มรายการในออเดอร์

        elif choice == "2":                  # ถ้าเลือก 2
            print("🧋 เมนูเครื่องดืม")
            print("น้ำเปล่า / ชาไทย / โอเลี้ยง / นมชมพู")
            drink = input("เลือกเครื่องดื่ม: ")
            
            if drink != "น้ำเปล่า":            # ถ้าไม่ใช่น้ำเปล่า
                sweet = input("ระดับความหวาน (น้อย / ปกติ / มาก): ")
            else:                           # ถ้าเลือกน้ำเปล่า
                sweet = "ไม่หวาน"
                
            cup = input("ขนาดแก้ว (เล็ก / ใหญ่): ")
            
            price, sweet = drink_price(drink, sweet, cup)                        # เรียก function คำนวณราคาเครื่องดื่ม
            
            orders.append(f"เครื่องดื่ม {drink} หวาน:{sweet} แก้ว:{cup} = {price}฿") # เพิ่มรายการในออเดอร์

        else:                              # ถ้าเลือกนอกเหนือจากนี้ 
            print("โปรดใส่ค่าตามตัวเลือก")

    # show order summary
    print("\n📦 สรุปออเดอร์")
    
    total = 0           # ประกาศตัวแปรเก็บ ราคารวมทุกรายการ
    
    # สรุปรายการทั้งหมด และราคารวม
    for i, order in enumerate(orders, start=1):
        print(f"{i}. {order}")
        total += int(order.split("=")[-1].replace("฿", "").strip())
    print(f"\n💵 รวม: {total} บาท")
# ----------------------------------------------------------------------------------------------------

main()              # เริ่มโปรแกรมทำงาน
```
