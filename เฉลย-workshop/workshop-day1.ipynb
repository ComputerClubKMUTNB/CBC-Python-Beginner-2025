### วิธีการคำนวณภาษี
1. ตรวจสอบรายได้ต่อปีของตัวเอง
2. คำนวณหารายได้ต่อปีสุทธิ โดยนำรายได้ต่อปีไปลดหย่อยภาษี (หักออก 60,000 บาท)
3. ตรวจสอบว่ารายได้ต่อปีสุทธิอยู่ที่ขั้นไหนของตาราง
4. ดูค่า จุดเริ่มต้นของขั้นนั้น, อัตราภาษีของขั้นนั้น และ ภาษีสูงสุดในแต่ละขั้นรายได้ของขั้นนั้น
5. ภาษี = (รายได้ต่อปีสุทธิ - จุดเริ่มต้นของขั้นนั้น) x อัตราภาษีของขั้นนั้น + ภาษีสูงสุดในแต่ละขั้นรายได้

**ตารางอัตราภาษีเงินได้บุคคลธรรมดาสำหรับปีภาษี 2560 ที่จะต้องยื่นรายการในปี พ.ศ.2561 เป็นต้นไป**



```python
salary = int(input("ใส่รายได้ต่อเดือน:"))                    # รายได้ต่อเดือน
income_per_year = salary * 12                                # รายได้ต่อปี   

print(f"รายได้ต่อเดือน = {salary} ,รายได้ต่อปี  = {income_per_year}")    # print ข้อมูลรายได้ต่อเดือนและรายได้ต่อปี
income_per_year -= 60_000                                           # รายได้ต่อปีสุทธิ เมื่อลดหย่อนภาษีแล้ว (รายได้ต่อปี - 60,000)

tax_rate = 0                            # ประกาศตัวแปรเก็บค่า อัตราภาษี
tax = 0                                 # ประกาศตัวแปรเก็บค่า ภาษีที่ต้องเสีย

# สูตรการคำนวณภาษี
# ภาษี = (รายได้ต่อปีสุทธิ - จุดเริ่มต้นของขั้นนั้น) x อัตราภาษีของขั้นนั้น + ภาษีสูงสุดในแต่ละขั้นรายได้

# คำถาม: ทำไมจากเงื่อนไขต้องเริ่มจากขั้นที่สูงที่สุด ?
# ตอบ : เพื่อป้องกัน input ที่ติดลบ
#       ในความเป็นจริงแล้ว input ที่ user ใส่ได้คือจำนวนเต็ม หมายความว่า user สามารถใส่ค่าติดลบได้
#       อาจจะแปลว่า user ติดหนี้! แน่นอนว่าเขาไม่จำเป็นต้องเสียภาษีด้วยซ้ำ
#       จากโค้ด เริ่มเงื่อนไขที่ขั้นที่สูงที่สุด (5,000,000) และให้เงื่อนไขสุดท้ายหรือ else ทำงานเมื่อ
#       user ใส่รายได้ที่อยู่ในช่วง 0 - 150,000 และรายได้ติดลบ

if income_per_year > 5_000_000:         # รายได้ต่อปีสุทธิมากกว่า 5,000,000 บาท
    # อัตราภาษี 35%
    tax_rate = 35                       
    tax = (income_per_year - 5_000_000) * (tax_rate / 100) + 1_265_000
    
elif income_per_year > 2_000_000:       # รายได้ต่อปีสุทธิมากกว่า 2,000,000 บาท
    # อัตราภาษี 30%
    tax_rate = 30
    tax = (income_per_year - 2_000_000) * (tax_rate / 100) + 365_000
    
elif income_per_year > 1_000_000:       # รายได้ต่อปีสุทธิมากกว่า 1,000,000 บาท
    # อัตราภาษี 25%
    tax_rate = 25
    tax = (income_per_year - 1_000_000) * (tax_rate / 100) + 11500
    
elif income_per_year > 750_000:         # รายได้ต่อปีสุทธิมากกว่า 750,000 บาท
    # อัตราภาษี 20%
    tax_rate = 20
    tax = (income_per_year - 750_000) * (tax_rate / 100) + 65000
    
elif income_per_year > 500_000:         # รายได้ต่อปีสุทธิมากกว่า 500,000 บาท
    # อัตราภาษี 15%
    tax_rate = 15
    tax = (income_per_year - 500_000) * (tax_rate / 100) + 27500
    
elif income_per_year > 300_000:         # รายได้ต่อปีสุทธิมากกว่า 300,000 บาท
    # อัตราภาษี 10%
    tax_rate = 10
    tax = (income_per_year - 300_000) * (tax_rate / 100) + 7500
    
elif income_per_year > 150_000:         # รายได้ต่อปีสุทธิมากกว่า 150,000 บาท
    # อัตราภาษี 5%
    tax_rate = 5
    tax = (income_per_year - 150_000) * (tax_rate / 100)
    
else:                                   # นอกเหนือจากนี้
    # อัตราภาษี 0 ไม่ต้องเสียภาษี
    tax_rate = 0
    tax = 0

print("อัตราภาษี", tax_rate)             # print อัตราภาษี
print("เสียภาษี =",tax)                  # print ภาษีที่ต้องจ่าย
```
