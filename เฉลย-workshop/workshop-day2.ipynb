```python
def noodle_price(sen, nam, nuea, size):         # function 
    price = 0                                   #   

    # 
    if sen == "" or sen == "":
        price += 0
    elif sen == "":
        price += 10
    elif sen == "":
        price += 15
    elif sen == "":
        price += 20

    # 
    if nam == "" or nam == "":
        price += 0
    elif nam == "":
        price += 10
    elif nam == "" or nam == "":
        price += 20

    #  / 
    if nuea == "":
        price += 10
    elif nuea == "":
        price += 20

    #  / 
    if size == "":
        price += 0
    elif size == "":
        price += 10

    return price                                    # return 
# ----------------------------------------------------------------------------------------------------

def drink_price(drink, sweet, cup):                 # function 
    price = 0                                       #   

    # 
    if drink == "" or drink == "":   #   
        price += 0
    elif drink == "":                    # 
        price += 20
    elif drink == "":                       # 
        price += 10

    # 
    if cup == "":                           # 
        price += 0
    elif cup == "":                         # 
        price += 10

    return price, sweet                             # return   
# ----------------------------------------------------------------------------------------------------

def main():                                         # function 
    orders = []                                     # 

    while True:                                     #  loop infinity
        print("\n: (1)  / (2)  / (0) ")
        choice = input(": ")      # 
        
        if choice == "0":                     #  0
            break                             #  loop 

        elif choice == "1":                   #  1
            print(" ")
            
            print(": ////")
            sen = input(": ")
            
            print(": ////")
            nam = input(": ")
            
            print(": /")
            nuea = input(": ")
            
            size = input(" (/): ")
            
            price = noodle_price(sen, nam, nuea, size)                      #  function 
            
            orders.append(f" {sen} {nam} {nuea} {size} = {price}")  # 

        elif choice == "2":                  #  2
            print(" ")
            print(" /  /  / ")
            drink = input(": ")
            
            if drink != "":            # 
                sweet = input(" ( /  / ): ")
            else:                           # 
                sweet = ""
                
            cup = input(" ( / ): ")
            
            price, sweet = drink_price(drink, sweet, cup)                        #  function 
            
            orders.append(f" {drink} :{sweet} :{cup} = {price}") # 

        else:                              #  
            print("")

    # show order summary
    print("\n ")
    
    total = 0           #  
    
    #  
    for i, order in enumerate(orders, start=1):
        print(f"{i}. {order}")
        total += int(order.split("=")[-1].replace("", "").strip())
    print(f"\n : {total} ")
# ----------------------------------------------------------------------------------------------------

main()              # 
```
