**ตารางราคา**



```python
def noodle_price(sen, nam, nuea, size):         # function คำนวณราคารายการก๋วยเตี๋ยว
    price = 0                                   # ประกาศตัวแปรเก็บ ราคาทั้งหมด ของรายการ

    # ประเภทเส้นก๋วยเตี๋ยว
    if sen == "เส้นเล็ก" or sen == "วุ้นเส้น":
        price += 0
    elif sen == "เส้นหมี่ขาว":
        price += 10
    elif sen == "เส้นใหญ่":
        price += 15
    elif sen == "เส้นหมี่เหลือง":
        price += 20

    # หมวดน้ำซุปก๋วยเตี๋ยว
    if nam == "แห้ง" or nam == "น้ำตก":
        price += 0
    elif nam == "น้ำใส":
        price += 10
    elif nam == "ต้มยำ" or nam == "เย็นตาโฟ":
        price += 20

    # หมวดหมูตุ๋น / เนื้อตุ๋น
    if nuea == "หมูตุ๋น":
        price += 10
    elif nuea == "เนื้อตุ๋น":
        price += 20

    # ธรรมดา / พิเศษ
    if size == "ธรรมดา":
        price += 0
    elif size == "พิเศษ":
        price += 10

    return price                                    # return ราคาทั้งหมดของรายการ
# ----------------------------------------------------------------------------------------------------

def drink_price(drink, sweet, cup):                 # function คำนวณราคารายการเครื่องดืม
    price = 0                                       # ประกาศตัวแปรเก็บ ราคาทั้งหมด ของรายการ

    # ประเภทเคครื่องดื่ม
    if drink == "ชาไทย" or drink == "โอเลี้ยง":   # ชาไทย หรือ โอเลี้ยง
        price += 0
    elif drink == "นมชมพู":                    # นมชมพู
        price += 20
    elif drink == "น้ำเปล่า":                       # น้ำเปล่า
        price += 10

    # ขนาดแก้ว
    if cup == "เล็ก":                           # แก้วเล็ก
        price += 0
    elif cup == "ใหญ่":                         # แก้วใหญ่
        price += 10

    return price, sweet                             # return ราคาทั้งหมดของรายการ และ ระดับบความหวาน
# ----------------------------------------------------------------------------------------------------

def main():                                         # function หลักที่จะรับออเดอร์
    orders = []                                     # ออเดอร์เก็บรายการอาหารที่สั่ง

    while True:                                     # วน loop infinity
        print("\nเลือกประเภทรายการ: (1) ก๋วยเตี๋ยว / (2) เครื่องดื่ม / (0) เพื่อปิดออเดอร์")
        choice = input("โปรดพิมพ์ตัวเลข: ")      # เลือกประเภทรายการด้วยตัวเลข
        
        if choice == "0":                     # ถ้าเลือก 0
            break                             # หยุด loop ปิดออเดอร์

        elif choice == "1":                   # ถ้าเลือก 1
            print("🍜 เมนูก๋วยเตี๋ยว")
            
            print("เส้น: เส้นเล็ก/วุ้นเส้น/เส้นหมี่ขาว/เส้นใหญ่/เส้นหมี่เหลือง")
            sen = input("เลือกเส้นก๋วยเตี๋ยว: ")
            
            print("น้ำซุป: แห้ง/น้ำใส/น้ำตก/ต้มยำ/เย็นตาโฟ")
            nam = input("เลือกน้ำซุป: ")
            
            print("เนื้อ: หมูตุ๋น/เนื้อตุ๋น")
            nuea = input("เลือกเนื้อ: ")
            
            size = input("เลือกขนาด (พิเศษ/ธรรมดา): ")
            
            price = noodle_price(sen, nam, nuea, size)                      # เรียก function คำนวณราคาก๋วยเตี๋ยว
            
            orders.append(f"ก๋วยเตี๋ยว {sen} {nam} {nuea} {size} = {price}฿")  # เพิ่มรายการในออเดอร์

        elif choice == "2":                  # ถ้าเลือก 2
            print("🧋 เมนูเครื่องดืม")
            print("น้ำเปล่า / ชาไทย / โอเลี้ยง / นมชมพู")
            drink = input("เลือกเครื่องดื่ม: ")
            
            if drink != "น้ำเปล่า":            # ถ้าไม่ใช่น้ำเปล่า
                sweet = input("ระดับความหวาน (น้อย / ปกติ / มาก): ")
            else:                           # ถ้าเลือกน้ำเปล่า
                sweet = "ไม่หวาน"
                
            cup = input("ขนาดแก้ว (เล็ก / ใหญ่): ")
            
            price, sweet = drink_price(drink, sweet, cup)                        # เรียก function คำนวณราคาเครื่องดื่ม
            
            orders.append(f"เครื่องดื่ม {drink} หวาน:{sweet} แก้ว:{cup} = {price}฿") # เพิ่มรายการในออเดอร์

        else:                              # ถ้าเลือกนอกเหนือจากนี้ 
            print("โปรดใส่ค่าตามตัวเลือก")

    # show order summary
    print("\n📦 สรุปออเดอร์")
    
    total = 0           # ประกาศตัวแปรเก็บ ราคารวมทุกรายการ
    
    # สรุปรายการทั้งหมด และราคารวม
    for i, order in enumerate(orders, start=1):
        print(f"{i}. {order}")
        total += int(order.split("=")[-1].replace("฿", "").strip())
    print(f"\n💵 รวม: {total} บาท")
# ----------------------------------------------------------------------------------------------------

main()              # เริ่มโปรแกรมทำงาน
```
